### Scrapping of bbc website with BeautifulSoup :

```python
# dependencies : 

import requests 

from bs4 import BeautifulSoup as bsp

import pandas as pd

```

```python
URL = 'http://feeds.bbci.co.uk/news/rss.xml'
```

```python
# send request to website 

response = requests.get(URL)

soup = bsp(response.text,'html.parser')
```

```python
# code html of page :
soup
```

```python
# inside item we have TITLE / PUBDATE / DESCRIPTION 

# We want to scrap these informations : 

items = soup.findAll('item')
```

```python
items
```

```python
items[0]
```

```python
# extract title : 

items[0].title.text
```

```python
# extract description :

items[0].description.text
```

```python
# extract pubdate :

items[0].pubdate.text
```

```python
items_=[]
for item in items:
    scrap={}
    
    scrap['title']=item.title.text
    scrap['description']=item.description.text
    scrap['pubdate']=item.pubdate.text
    
    items_.append(scrap)
```

```python
items_
```

```python
# save data into dataframe with pandas : 

bbc_news = pd.DataFrame(items_,columns=['title','description','pubdate'])
```
